```python
import pandas as pd
import numpy as np
Filename = 'SparklingDB'
df = pd.read_excel(Filename+'.xlsx')
print('결측치 처리하기 전')
print(df)
```

```python
#----------------결측치 처리-----------------#
#=> 결측치 처리한 파일을 마지막 DB로 사용할 것이며, 이미지를 다운로드 받은 후, CNN모델을 만들고 Transfer learning하여 모델을 생성
df = df.dropna(subset=['img'])
df['expert_comment']=df['expert_comment'].fillna(0)
df['review']=df['review'].fillna(0)
df['country_img']=df['country_img'].fillna(0)
df['wine_label1']=df['wine_label1'].fillna(0)
df['wine_label2']=df['wine_label2'].fillna(0)
print(df)
#--------------------------------------------#
```

```python
print('결측치 처리 한 파일 저장')
df.to_excel(Filename+'_Final.xlsx')
```

```python
print('이미지 다운로드')
import urllib.request
for i in range(0,1891):                           
    f=open('./img/wine'+str(i)+'.jpg','wb')
    a = df.iloc[i][0]
    f.write(urllib.request.urlopen(a).read())
    f.close()
```
